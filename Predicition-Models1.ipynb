```python
city_data  = data.iloc[:,14:]
city_data
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
country = ohe.fit_transform(city_data).toarray()
country
```

```python
column_names = ohe.get_feature_names_out(['city', 'country'])
df_encoded = pd.DataFrame(country, columns=column_names)
data = pd.concat([data, df_encoded], axis=1)

data.head()
```
