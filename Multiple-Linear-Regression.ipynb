```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('datas.csv',sep=',')
data.head()
```

```python
temp_data = data.values
```

```python
country = data.iloc[:,0:1].values
country
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
country[:,0] = le.fit_transform(data.iloc[:,0])
country
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
country = ohe.fit_transform(country).toarray()
country
```

```python
gender = data.iloc[:,-1:].values
gender[:,-1] = le.fit_transform(data.iloc[:,-1])
gender = ohe.fit_transform(gender).toarray()
gender
```

```python
s1 = pd.DataFrame(data = country, index = range(22), columns=['fr','tr','us'] )
s2 = pd.DataFrame(data = gender[:,:1] , index=range(22), columns=['gender'])
s3 = data.iloc[:,1:4]
```

```python
d1=pd.concat([s1,s3],axis=1)
data2=pd.concat([d1,s2],axis=1)
data2
```

```python
x_data = data2.iloc[:,:-1]
x_data
```

```python
y_data = data2.iloc[:,-1:]
y_data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.33, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train,y_train)
```

```python
y_pred = lin_reg.predict(x_test) #tahmin
print(y_pred)
print(y_test)
```

```python
height = data2.iloc[:,3:4].values
height
```

```python
t1 = data2.iloc[:,:3]
t2 = data2.iloc[:,4:]
data3 = pd.concat([t1,t2],axis=1)
data3
```

```python
x_train, x_test,y_train,y_test = train_test_split(data3,height,test_size=0.33, random_state=0)
```

```python
y_pred = lin_reg.predict(x_test) #tahmin
print(y_pred)
print(y_test)
```
